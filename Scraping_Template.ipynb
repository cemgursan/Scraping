```python
#Importing pips
import bs4
from urllib.request import urlopen 
from bs4 import BeautifulSoup
from openpyxl import Workbook
import datetime
import pandas as pd
from pandas import DataFrame, Series
```

```python
#Excel File
wb = Workbook()
ws = wb.active
headers=["News Header","News Description","News Time","News Type/Location"]
NowTag=str(datetime.datetime.now())
ws.append([NowTag])
ws.append(headers)
```

```python
# Connecting to the Web Site

UrlToOpen ="https://www.bbc.com/news"  # Change Link if needed!
UrlOpened = urlopen(UrlToOpen)

# Dumping the website into specified variables

htmlpage = UrlOpened.read()

# Close Connection

UrlOpened.close()

#Html Parsing

htmlsoup = BeautifulSoup(htmlpage,"html.parser")
```

```python
#Scraping necessary parts

NewsDump = htmlsoup.findAll("div",{"class":"gs-c-promo-body gel-1/2@xs gel-1/1@m gs-u-mt@m"})
#Lets loop it
#NewsLoop = NewsDump[5] #Tutorial DELETE IT
#NewsLoop.h3

counter=0
for NewsLoop in NewsDump:
    
    #News Header
    NewsHeader = NewsLoop.h3.text
    
    #News Paragraph
    if NewsLoop.p != None:
        NewsDesc = NewsLoop.p.text       
    #Some News doesnt have Paragraph 
    else:
        NewsDesc = "No Description Avalaible"
        
    #News Time
    try:
        NewsTime = NewsLoop.ul.li.time["data-datetime"]    #News Time
    #Some News doesnt have Time
    except:
        NewsTime = "No Time Avalaible"
       
        
    try: 
        TempLoc = NewsLoop.findAll("li",{"class":"nw-c-promo-meta"})
        NewsType = TempLoc[1].span.text     #News Location
    except:
        NewsType = "No Location or News Type Specified"
    
    ws.append([NewsHeader,NewsDesc,NewsTime,NewsType])

```

```python
data = pd.read_excel('BBC News Scraping Tutorial.xlsx')
df1=DataFrame(data)
#Delete the time information above
df1.columns=df1.loc[0]
#Drop the duplicate
df1=df1.drop(df1.index[0])
df1
```

```python
#Let's find interesting news in the list
wordtosearch ="S"
rownumbers=[]
counter=0
resultframe= pd.DataFrame()
print(resultframe)
#Search for the word in the dataframe and save the found number in rownumbers
for row in df1["News Header"]:
    result = row.find(wordtosearch)
    
    if result>-1:
        rownumbers.append(counter)
    
    counter=counter+1    
    
#Let's print the found data 
for item in rownumbers:

    tempdata=df1.iloc[item].to_frame().T
    resultframe=resultframe.append(tempdata)
    
    
resultframe
```
